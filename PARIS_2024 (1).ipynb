```python
# Dados sobre a história dos Jogos Olímpicos
jogos_olimpicos_historia = {
    "Jogos Olímpicos Antigos": {
        "Origem": "Grécia Antiga, aproximadamente 776 a.C.",
        "Eventos Originais": ["Corrida de pé (stade)"],
        "Evolução": [
            "Luta livre",
            "Pentatlo (corrida, salto em distância, arremesso de disco, lançamento de dardo e luta)",
            "Corridas de cavalos e carruagens",
            "Pancrácio (combinação de luta livre e boxe)"
        ],
        "Declínio": "Proibidos em 393 d.C. pelo imperador Teodósio I"
    },
    "Jogos Olímpicos Modernos": {
        "Renascimento": {
            "Fundador": "Barão Pierre de Coubertin",
            "Comitê Olímpico Internacional (COI)": "Fundado em 1894",
            "Primeiros Jogos": {
                "Ano": 1896,
                "Cidade": "Atenas, Grécia",
                "Participantes": {"Atletas": 241, "Países": 14}
            }
        },
        "Desenvolvimento": {
            "Crescimento": [
                "Número de esportes e eventos aumentou",
                "Realização a cada quatro anos, com exceção de guerras"
            ]
        },
        "Inclusão de Novos Esportes e Eventos": {
            "Exemplos": [
                "Basquete",
                "Vôlei",
                "Judô",
                "Skateboarding",
                "Surfe",
                "Escalada esportiva"
            ],
            "Jogos de Inverno": {
                "Início": 1924,
                "Cidade": "Chamonix, França"
            }
        },
        "Aspectos Culturais e Políticos": [
            "Boicotes de 1980 e 1984",
            "Protestos de 1968 na Cidade do México",
            "Ataque terrorista em Munique em 1972"
        ],
        "Jogos Paralímpicos": {
            "Início": 1960,
            "Cidade": "Roma, Itália"
        },
        "Últimos Jogos": {
            "Verão 2020": {"Ano": 2021, "Cidade": "Tóquio, Japão"},
            "Inverno 2022": {"Ano": 2022, "Cidade": "Pequim, China"},
            "Próximos Jogos Verão 2024": {"Ano": 2024, "Cidade": "Paris, França"}
        }
    }
}

# Função para exibir a história dos Jogos Olímpicos
def exibir_historia_jogos_olimpicos(historia):
    for periodo, dados in historia.items():
        print(f"{periodo}:")
        if isinstance(dados, dict):
            for chave, valor in dados.items():
                if isinstance(valor, dict):
                    print(f"  {chave}:")
                    for sub_chave, sub_valor in valor.items():
                        if isinstance(sub_valor, dict):
                            print(f"    {sub_chave}:")
                            for item_chave, item_valor in sub_valor.items():
                                print(f"      {item_chave}: {item_valor}")
                        else:
                            print(f"    {sub_chave}: {sub_valor}")
                else:
                    print(f"  {chave}: {valor}")
        print()

# Executa a função para exibir a história dos Jogos Olímpicos
exibir_historia_jogos_olimpicos(jogos_olimpicos_historia)
```

```python
#Apresentação das medalhas e dos atletas.
import pandas as pd
import matplotlib.pyplot as plt

# Dados atualizados
dados = {
    'Atleta': ['Beatriz Souza', 'Rebeca Andrade', 'Ana Patrícia e Duda', 'Caio Bonfim', 'Willian Lima',
              'Rebeca Andrade', 'Rebeca Andrade', 'Tatiana Weston-Webb', 'Isaquias Queiroz', 'Brasil',
              'Larissa Pimenta', 'Rayssa Leal', 'Brasil', 'Brasil', 'Bia Ferreira', 'Gabriel Medina',
              'Augusto Akio', 'Edival Pontes "Netinho"', 'Alison dos Santos', 'Brasil'],
    'Medalha': ['Ouro', 'Ouro', 'Ouro', 'Prata', 'Prata',
               'Prata', 'Prata', 'Prata', 'Prata', 'Prata',
               'Bronze', 'Bronze', 'Bronze', 'Bronze', 'Bronze', 'Bronze',
               'Bronze', 'Bronze', 'Bronze', 'Bronze']
}
df = pd.DataFrame(dados)

# Definir cores
cores = {'Ouro': 'gold', 'Prata': 'silver', 'Bronze': 'peru'}

# Mapear medalhas para valores numéricos
df['Valor Medalha'] = df['Medalha'].map({'Ouro': 3, 'Prata': 2, 'Bronze': 1})

# Ordenar os atletas pelo valor da medalha
df_sorted = df.sort_values(by='Valor Medalha', ascending=False)

# Plotar o gráfico
plt.figure(figsize=(14, 8))
bars = plt.bar(df_sorted['Atleta'], df_sorted['Valor Medalha'], color=df_sorted['Medalha'].map(cores), edgecolor='black')

# Adicionar anotações com o tipo de medalha
for bar, medalha in zip(bars, df_sorted['Medalha']):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, medalha, ha='center', va='bottom', fontsize=10, color='black')

# Melhorar layout do gráfico
plt.title('Distribuição de Medalhas por Atleta', fontsize=16, fontweight='bold')
plt.xlabel('Atleta', fontsize=14)
plt.ylabel('Valor da Medalha', fontsize=14)
plt.xticks(rotation=90, ha='right', fontsize=12)
plt.yticks([1, 2, 3], ['Bronze', 'Prata', 'Ouro'], fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
