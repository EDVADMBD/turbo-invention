```python
#Retornar os dados em um gráfico da Ação de Vendas Peugeot Periodo de Jan a Maio 2024
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import random

class PeugeotSalesAction:
    def __init__(self):
        self.customers = []
        self.potential_customers = ["Cliente1", "Cliente2", "Cliente3", "Cliente4", "Cliente5"]
        self.new_models = ["Peugeot 208", "Peugeot 2008", "Peugeot 3008"]
        self.actions = []
        self.dates = []

    def add_action(self, action, date):
        self.actions.append(action)
        self.dates.append(date)

    def offer_promotions(self, date):
        promo = "Manutenção gratuita por 1 ano"
        for customer in self.potential_customers:
            action = f"Oferecendo {promo} para {customer}"
            self.add_action(action, date)
        self.customers.extend(self.potential_customers)
        self.potential_customers = []

    def launch_new_models(self, date):
        for model in self.new_models:
            action = f"Lançando novo modelo: {model}"
            self.add_action(action, date)

    def run_digital_marketing_campaign(self, date):
        action = "Executando campanha de marketing digital..."
        self.add_action(action, date)
        new_leads = ["Lead1", "Lead2", "Lead3"]
        self.potential_customers.extend(new_leads)

    def offer_test_drives(self, date):
        for customer in self.potential_customers:
            action = f"Oferecendo test drive para {customer}"
            self.add_action(action, date)

    def build_partnerships(self, date):
        action = "Construindo parcerias com influenciadores e outras empresas..."
        self.add_action(action, date)

    def participate_in_events(self, date):
        action = "Participando de eventos e feiras automotivas..."
        self.add_action(action, date)

    def implement_loyalty_program(self, date):
        action = "Implementando programa de fidelidade para clientes recorrentes..."
        self.add_action(action, date)

    def improve_customer_service(self, date):
        action = "Melhorando a experiência de atendimento ao cliente..."
        self.add_action(action, date)

    def execute_sales_action(self):
        start_date = datetime(2024, 1, 1)
        end_date = datetime(2024, 5, 31)
        date_range = pd.date_range(start_date, end_date).to_pydatetime().tolist()

        for date in date_range:
            action_day = random.choice([
                self.launch_new_models,
                self.run_digital_marketing_campaign,
                self.offer_promotions,
                self.offer_test_drives,
                self.build_partnerships,
                self.participate_in_events,
                self.implement_loyalty_program,
                self.improve_customer_service
            ])
            action_day(date)

        self.add_action("Ação de vendas executada com sucesso!", end_date)

    def get_actions_dataframe(self):
        data = {
            "Data": self.dates,
            "Ação": self.actions
        }
        df = pd.DataFrame(data)
        return df

# Execute a ação de vendas da Peugeot
peugeot_sales = PeugeotSalesAction()
peugeot_sales.execute_sales_action()

# Obter o dataframe das ações realizadas
df_actions = peugeot_sales.get_actions_dataframe()

# Filtrar o dataframe para o período de janeiro a maio de 2024
start_date = "2024-01-01"
end_date = "2024-05-31"
mask = (df_actions['Data'] >= start_date) & (df_actions['Data'] <= end_date)
df_filtered = df_actions.loc[mask]

# Contagem das ações realizadas no período
actions_count = df_filtered['Ação'].value_counts()

# Plotagem do gráfico
plt.figure(figsize=(10, 6))
actions_count.plot(kind='barh', color='skyblue')
plt.xlabel('Quantidade')
plt.ylabel('Ações')
plt.title('Análise das Ações de Vendas da Peugeot (Jan-Mai 2024)')
plt.show()


```

```python
#simulação onde os carros da Peugeot têm diferentes volumes de vendas fictícios, preços e ofertas de financiamento
import pandas as pd
import matplotlib.pyplot as plt

# Dados fictícios
carros = ['Peugeot 208', 'Peugeot 2008', 'Peugeot 3008']
vendas = [120, 90, 150]
precos = [65000, 85000, 120000]
financiamentos = [0, 2, 1]  # Taxa de juros em %

# Criando um DataFrame com os dados
df_carros = pd.DataFrame({
    'Modelo': carros,
    'Vendas': vendas,
    'Preço (R$)': precos,
    'Financiamento (%)': financiamentos
})

# Encontrando o carro mais vendido
carro_mais_vendido = df_carros.loc[df_carros['Vendas'].idxmax()]

# Encontrando o carro menos vendido
carro_menos_vendido = df_carros.loc[df_carros['Vendas'].idxmin()]

# Encontrando o carro mais caro
carro_mais_caro = df_carros.loc[df_carros['Preço (R$)'].idxmax()]

# Encontrando o carro mais barato
carro_mais_barato = df_carros.loc[df_carros['Preço (R$)'].idxmin()]

# Encontrando o carro com melhor financiamento (menor %)
carro_melhor_financiamento = df_carros.loc[df_carros['Financiamento (%)'].idxmin()]

# Encontrando o carro com pior financiamento (maior %)
carro_pior_financiamento = df_carros.loc[df_carros['Financiamento (%)'].idxmax()]

# Preparando dados para o gráfico
dados = {
    'Modelo': [
        carro_mais_vendido['Modelo'], carro_menos_vendido['Modelo'],
        carro_mais_caro['Modelo'], carro_mais_barato['Modelo'],
        carro_melhor_financiamento['Modelo'], carro_pior_financiamento['Modelo']
    ],
    'Valor': [
        carro_mais_vendido['Vendas'], carro_menos_vendido['Vendas'],
        carro_mais_caro['Preço (R$)'], carro_mais_barato['Preço (R$)'],
        carro_melhor_financiamento['Financiamento (%)'], carro_pior_financiamento['Financiamento (%)']
    ]
}

df_dados = pd.DataFrame(dados)

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.barh(df_dados['Modelo'], df_dados['Valor'], color='skyblue')
plt.xlabel('Valor')
plt.title('Comparação entre Carros da Peugeot')
plt.gca().invert_yaxis()  # Inverte a ordem das barras para a visualização mais vendido para o menos vendido
plt.show()
```
